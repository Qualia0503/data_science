```python
## 载入需要的包

from folium import plugins
import folium
import pandas as pd
```

```python
df = pd.read_csv("东风商用车(合肥市).csv",header=0,encoding='utf-8')
df.columns = ["SUOYIN","NAME","TYPE","ADDRESS","QUYU","WEI","JING"]
df['BRANDNAME'] = df.NAME.apply(lambda x: 1 if "东风商用" in x else 0)
# df['LJ'] = df.NAME.apply(lambda x: 1 if '链家' in x else 0)

df_BRANDNAME = df[(df["BRANDNAME"]==1)].reset_index(drop=True)
# df_lj = df[(df["LJ"]==1)].reset_index(drop=True)

ss1 = [(lat, lon) for lat, lon in zip(df_BRANDNAME["JING"],df_BRANDNAME["WEI"])]
# ss2 = [(lat, lon) for lat, lon in zip(df_lj["JING"],df_lj["WEI"])]


m = folium.Map([31.820567, 117.227267], # 该市中心区域经纬度
    zoom_start=10,
                tiles='http://webrd02.is.autonavi.com/appmaptile?lang=zh_cn&size=1&scale=1&style=7&x={x}&y={y}&z={z}', # 高德街道图
                # tiles='https://mt.google.com/vt/lyrs=h&x={x}&y={y}&z={z}', # google 地图
                # tiles='Stamen Toner',
               attr='高德')  

# 绘制店标记
groups = folium.FeatureGroup('')

# for n,label in zip(ss2,df_lj['NAME'].to_list()):
#     groups.add_child(
#         folium.CircleMarker(
#             location=n,
#             radius=3,
#             color='green',
#             fill=True,
#             fill_color='green',
# #             fill_opacity=0.7,
#             popup=folium.Popup(html=label,max_width=100),
#         )
#     )


for l,label in zip(ss1,df_BRANDNAME['BRANDNAME'].to_list()):
    groups.add_child(
        folium.CircleMarker(
            location=l,
            radius=8,
            color='yello',
            fill=True,
            fill_color='yellow',
            fill_opacity=0.4,
            popup=folium.Popup(html=label,max_width=100),
        )
    )

    
    
m.add_child(groups)
m.add_child(folium.LatLngPopup())  

# 区域密度
marker_cluster = plugins.MarkerCluster().add_to(m)      

# for n,label in zip(ss2,df_lj['NAME'].to_list()):
#     # 添加marker到地图
#     folium.Marker(
#         location = n,
#         icon=None,
#         # popup=folium.Popup(html=label,max_width=100),
#     ).add_to(marker_cluster)

for l,label in zip(ss1,df_BRANDNAME['BRANDNAME'].to_list()):
    # 添加marker到地图
    folium.Marker(
        location = l,
        icon=None,
        # popup=folium.Popup(html=label,max_width=100),
    ).add_to(marker_cluster)

        
m.add_child(marker_cluster)

m.save('hefei_dongfeng.html')  # 将结果以HTML形式保存到本地
```
