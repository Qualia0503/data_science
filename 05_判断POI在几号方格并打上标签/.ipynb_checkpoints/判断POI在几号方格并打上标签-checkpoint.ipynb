```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from tqdm.notebook import tqdm_notebook
tqdm_notebook.pandas()
```

```python
point = pd.read_excel('beijingpoichooseVZ2.xlsx',sheet_name="A")

grid = pd.read_excel('beijingpoichooseVZ2.xlsx',sheet_name="C")
```

```python
def find_grid(lng,lat):
    
    global grid
    
    c1=grid['lng_left']<=lng
    c2=grid['lng_right']>lng
    c3=grid['lat_left']<=lat
    c4=grid['lat_right']>lat
    
    try:
        find = grid[c1&c2&c3&c4]['FID'].iat[0]
    except:
        find = 'error'
    
    return find
```

```python
# find_grid(115.744135,39.735088)
point_square = point[:]
point_square
```

```python
point_square['FID0']=point_square.progress_apply(lambda x: find_grid(x['WGS84经度'],x['WGS84纬度']),axis=1)


# point_square.to_excel('points with grid1.xlsx')
```

```python
point_square.to_csv('points_with_grid0908.csv',header=True,index=False)
```

```python
df_points  = pd.read_csv("points_with_grid0908.csv",header=0,encoding='utf-8')
df_points
```

```python
df_points["名称"].value_counts()
```

```python
df_points["名称"].fillna(value="这是空值",inplace=True)
df_points
```

```python
poiname = df_points.groupby('FID0').agg({
    '名称':[','.join]
}).reset_index()
poiname.columns = ["FID0","名称"]
poiname
```

```python
poiname['tonghang1'] = poiname["名称"].str.count("火锅")
poiname['tonghang2'] = poiname["名称"].str.count("小肥羊")
poiname['tonghang3'] = poiname["名称"].str.count("呷哺")
poiname['tonghang4'] = poiname["名称"].str.count("局气")
poiname['tonghang5'] = poiname["名称"].str.count("受气")
# poiname['tonghang6'] = poiname["名称"].str.count("潮汕")
poiname['tonghang'] = poiname['tonghang1']+poiname['tonghang2']+poiname['tonghang3']+poiname['tonghang4']+poiname['tonghang5']
poiname
```

```python
poiname.to_excel("poiname_tonghang.xlsx",header=True,index=False)
```

```python
df_points_unstack = df_points["中类"].groupby(df_points["FID0"]).value_counts()
df_points_unstack = df_points_unstack.unstack()
df_points_unstack = df_points_unstack.reset_index()
df_points_unstack
```

```python
df_points_unstack.to_excel("df_points_unstack21.xlsx",header=True,index=False)
```
